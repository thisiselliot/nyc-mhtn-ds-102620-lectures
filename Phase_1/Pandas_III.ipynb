## Part II. Pandas Groupby
Groupby methods in Pandas allow you to aggregate data and perform operations on them. The method can be summarized as split-apply-combine. The groupby function allow us to split the dataframe into _groups_, apply operations on them, and aggregate a final result. Let's look at some examples. 

```python
'''we have been looking at summary stats
on an aggregate lvl. But what if we want to calc
summary stats for each category ... split my
dataset up such that countries that belong to
africa, I want to calc their mean and med sep
from countries belonging to EU'''
```

```python
'''that is the idea of group by.

group by splits up your obs according to some
grouping var, applies an operation to it,
and then combines for a final result.

So group by allows us to split the DF into groups,
apply operations to them, then agregate a final
result'''
```

```python
'''Some examples...'''
```

```python
'''Group by continents, find out is there
higher beer consumption for certain continents
compared to others. So, in ordeer to review summary
statistics for every continent what is the grouping
ver here, what do I want to group it by?'''
```

```python
'''result of a groupby(object) is a groupby_object
you need to perform operation on the result ... 
group them by continent to see summary stats per
continent ... so finish up with .describe()'''

df.groupby(['continent']).describe()
```

```python
'''you can also pass this to a var'''

df.groupby('continent')['beer_servings'].mean()
```

```python
df.groupby('continent')['beer_servings'].mean().sort_values()
```

```python
# review summary statistics for each continent
var = df.groupby('continent')['beer_servings']

var.mean()
```

```python
'''there you go ... data analysis !!! '''
```

```python
# which continent drinks more beer on average?
'''perform operation on a certain column after
doing df.groupby()

the var of interest is continent, what is column
of interest, that we want to perform operations on

^^^beer servings <-- 

either .notation() or ['sq_bracket']'''

df.groupby(
    'continent'
)['beer_servings'].mean().sort_values()
```

```python
'''whatever you group it by is going to
be the index of the result'''

df.groupby(
    'continent'
)[
    'beer_servings'
].mean().sort_values(ascending=False).index
```

```python
'''what happens if we group by multiple
indices <-- that is when we get a
*multi-level series*'''
```

```python
# exercise - which continent drinks more beer than wine?
'''*** There is many ways to solve this prob. You
can also just find w/ .descrobe(), but what are the
other queries to think about this question'''
```

```python
'''what if we add .max() at the end?'''

'''this just returns the result without
an index, what country is it?'''
df.groupby(
    ['continent']
)['beer_servings'].mean().max()
```

```python
'''better to do a sort_values(ascending=False)'''

df.groupby(
    ['continent']
)['beer_servings'].mean().sort_values(
ascending = False)
```

```python
df.groupby(
    ['continent']
)['beer_servings'].mean().sort_values(
ascending = False)[:1]
```

```python
'''what is the grouping var if I want
to know mean age per occupation -- 
occupation is the grouping var'''

'''if I want to find mean age per
occupation, what is the column to which
I want to apply operation? Age. And the
operation to which I want to apply to it is
mean()'''
```

```python
# calculate the mean age per occupation
occupations.groupby(['occupation']).age.mean()
```

```python
# for each combination of sex and
#  occupation, show the mean age,
#  assigned it to group by age
'''so there are 2 grouping vars'''

occupations.groupby(
    ['gender', 'occupation']
).age.mean()

# This is a series, not a DF, it is a
# Multilevel Series, so still
# 1 dimensional, no column names

```

```python
group_by_age = occupations.groupby(
    ['occupation', 'gender']
).age.mean()

type(group_by_age)
```

```python
# exercise - for each occupation, calculate the percentage of female
# and sort it from high to low

# occupations.groupby(['occupation'])
#   ['gender'=='F'].value_counts
#   (normalize=True).sort_values()

# occupations[occupations['gender']=='F']

# hint use value counts
# Normalize=True
occupations.groupby(
    ['occupation']
).gender.value_counts(
    normalize=True
)[:,'F'].sort_values(
    ascending=False
) * 100

# .value_counts() is a method for DF's &
# Series, NOT GroupBy Objects,  cannot be applied
#  to a GroupBY Obj.
```

```python
''' ^^^ Use normalize to get a percentage. Use
bracket notation to select only the female
([:,'F']) and then use sort_values(ascending=
False) to sort from high to low, and multiply
by 100 to get percentage'''
```

```python
'''What is allowing us to use [,'F'] without
using .loc or .iloc? Because this is a GroupBy
Obj not a Series and GroupBy Obj's do not req.
.loc or .iloc to index by row and column.'''
```

```python
# if you want all the levels to be together, just sort the values
occupations.sort_values(by='gender')
```
